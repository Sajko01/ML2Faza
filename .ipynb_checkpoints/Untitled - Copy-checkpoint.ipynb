## Detekcija i obrada outlier-a

Outlier-i su vrednosti koje značajno odstupaju od ostatka podataka. U kontekstu
klasterizacije, oni mogu imati sledeće negativne efekte:

- distorziraju udaljenost između instanci,
- utiču na centroid-e kod K-Means algoritma,
- mogu dovesti do stvaranja dodatnog klastera koji nije smislen,
- otežavaju interpretaciju rezultata.

Zbog toga je identifikacija i odgovarajuća obrada outlier-a ključna za kvalitet
dobijenih klastera.


Za detekciju outlier-a koristićemo **IQR metodu**, jer je robustna prema
asimetričnim distribucijama, što je slučaj u ovom dataset-u. Z-score metoda
se može koristiti kao dodatna provera.


```python
# Funkcija za detekciju outlier-a
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Provera za najkritičnije numeričke atribute
critical_features = ['Age', 'Average_Order_Value', 'Lifetime_Value', 'Cart_Abandonment_Rate']

for feature in critical_features:
    outliers = detect_outliers_iqr(df_selected, feature)
    print(f"{feature}: {len(outliers)} outlier-a ({len(outliers)/len(df_selected)*100:.2f}%)")
```

IQR analiza pokazuje da su najkritičniji atributi sa najvećim brojem outlier-a:

- *Age* → nekoliko stotina ekstremnih vrednosti (npr. 200 godina)
- *Average_Order_Value* → nekoliko stotina veoma visokih kupovina
- *Lifetime_Value* → ekstremne vrednosti preko 8000
- *Cart_Abandonment_Rate* → vrednosti preko 100%

Ovi outlier-i mogu negativno uticati na klasterizaciju i moraju se obraditi.


Za ovaj projekat odlučeno je da se primeni **capping metoda (winsorizing)**, koja
ograničava ekstremne vrednosti na donju i gornju granicu IQR-a. Na ovaj način
se zadržavaju sve instance, a efekat outlier-a na formiranje klastera se smanjuje.


```python
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# Primena capping-a na kritične feature-e
for feature in critical_features:
    df_selected = cap_outliers_iqr(df_selected, feature)
```

```python
for feature in critical_features:
    outliers = detect_outliers_iqr(df_selected, feature)
    print(f"{feature} after capping: {len(outliers)} outlier-a ({len(outliers)/len(df_selected)*100:.2f}%)")
```

Primena capping metode omogućava očuvanje svih instanci u dataset-u, dok se
istovremeno eliminiše negativni uticaj ekstremnih vrednosti na algoritme
zasnovane na distancama. Ova metoda je posebno pogodna za klasterizaciju i
omogućava dobijanje stabilnijih i interpretabilnih klastera.


## Redukcija dimenzionalnosti

Skup podataka nakon selekcije i kodiranja sadrži više od 20 atributa.
Algoritmi za klasterizaciju, naročito vizuelizacija rezultata, postaju
teško interpretabilni u visoko-dimenzionalnom prostoru.

Redukcija dimenzionalnosti omogućava:

- vizuelizaciju podataka u 2D ili 3D prostoru,
- smanjenje šuma i redundantnih informacija,
- ubrzanje algoritama za klasterizaciju,
- bolju interpretaciju dobijenih klastera.

Najčešće korišćene metode su:
- **PCA (Principal Component Analysis)** – linearna metoda, fokus na varijansu
- **t-SNE (t-Distributed Stochastic Neighbor Embedding)** – nelinearna metoda, fokus na očuvanje lokalnih odnosa


```python
from sklearn.decomposition import PCA

# Kreiramo PCA objekat za 2 komponente
pca = PCA(n_components=2)

# Primena PCA na skalirane podatke
X_pca = pca.fit_transform(X_scaled)

# Pregled oblika podataka
X_pca.shape
```

```python
explained_variance = pca.explained_variance_ratio_
print(f"PC1: {explained_variance[0]*100:.2f}% varijanse")
print(f"PC2: {explained_variance[1]*100:.2f}% varijanse")
```

Prve dve PCA komponente objašnjavaju značajan deo varijanse u podacima.
Ovo nam omogućava da 2D vizualizacijom dobijemo realan pregled strukture
podataka i potencijalnih klastera.


```python
plt.figure(figsize=(10, 7))
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=10, alpha=0.6)
plt.title('2D PCA projection of E-commerce customers')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

Vizualizacija PCA pokazuje grupisanje instanci u prostoru prvih dve komponente.
Još ne znamo klastere, ali možemo uočiti potencijalne zone gde se podaci
koncentrišu, što je korisno za inicijalnu procenu broja klastera.


```python
pca_full = PCA()
pca_full.fit(X_scaled)

plt.figure(figsize=(8,5))
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.title('Explained Variance by PCA Components')
plt.grid()
plt.show()
```

Na osnovu kumulativne varijanse može se videti da je za očuvanje 80% varijanse potrebno 15+ komponenti, dok su dve komponente korišćene isključivo za vizualizaciju.

t-SNE je metoda koja bolje očuvava lokalne odnose među podacima, odnosno
slični podaci ostaju bliže, dok se različiti podaci udaljavaju. 
Posebno je korisna za vizualizaciju visokodimenzionalnih podataka.


```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

tsne = TSNE(n_components=2, perplexity=30, max_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(10, 7))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], s=10, alpha=0.6)
plt.title('2D t-SNE projection of E-commerce customers')
plt.show()
```

t-SNE vizualizacija često otkriva finije strukture u podacima i može pomoći
u izboru broja klastera i identifikaciji potencijalnih izolovanih grupa.


Redukcija dimenzionalnosti omogućava:

- bržu i stabilniju klasterizaciju,
- vizuelnu procenu strukture podataka,
- identifikaciju potencijalnih izolovanih instanci ili “outlier-a”.

PCA je koristan za globalne obrasce, dok t-SNE pomaže u otkrivanju lokalnih
grupa i finijih obrazaca ponašanja korisnika.


## Primena algoritama za klasterizaciju

Korišćenje više algoritama omogućava:

- upoređivanje dobijenih klastera,
- ocenu kvaliteta klastera po različitim kriterijumima,
- identifikaciju optimalnog broja klastera,
- razumevanje kako različite metode obrađuju gustinu i distribuciju podataka.

Najčešće korišćeni algoritmi u ovom projektu:

1. K-Means (linearna klasterizacija zasnovana na centroidima)
2. DBSCAN (detekcija gustih regiona i outlier-a)
3. Hijerarhijska klasterizacija (opcioni korak)


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []
K_range = range(2, 11)  # testiramo K od 2 do 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Vizualizacija Elbow metode
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, 'bo-')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import calinski_harabasz_score

K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    
    silhouette = silhouette_score(X_scaled, labels)
    davies_bouldin = davies_bouldin_score(X_scaled, labels)
    calinski = calinski_harabasz_score(X_scaled, labels)
    
    print(f"K={k}")
    print(f"  Silhouette Score: {silhouette:.4f}")
    print(f"  Davies-Bouldin Index: {davies_bouldin:.4f}")
    print(f"  Calinski-Harabasz Score: {calinski:.2f}")
    print("-" * 40)
```

## Evaluacija KMeans klastera

Kvalitet klastera procenjen je korišćenjem tri interne metrike:

- **Silhouette Score** – meri koheziju unutar klastera i separaciju između klastera. Vrednosti su u opsegu od -1 do 1, pri čemu veće vrednosti ukazuju na bolju razdvojenost klastera.
- **Davies–Bouldin Index (DBI)** – meri prosečnu sličnost između klastera. Niže vrednosti ukazuju na bolju separaciju i kompaktnije klastere.
- **Calinski–Harabasz Score (CH)** – odnos između međuklasterske disperzije i unutarklasterske disperzije. Veće vrednosti ukazuju na bolje definisane klastere.

### Analiza rezultata

Najbolji rezultati postignuti su za **K=2**:

- Silhouette Score: **0.2250** (najviša vrednost)
- Davies–Bouldin Index: **1.6651** (najniža vrednost)
- Calinski–Harabasz Score: **15399.62** (najviša vrednost)

Povećanjem broja klastera dolazi do opadanja Silhouette i Calinski–Harabasz vrednosti, dok Davies–Bouldin indeks uglavnom raste, što ukazuje na slabiju separaciju i veću fragmentaciju klastera.

Na osnovu kombinacije sva tri kriterijuma može se zaključiti da je **K=2 optimalan broj klastera**, jer pruža najbolju ravnotežu između kohezije i separacije u odnosu na ostale testirane vrednosti K.

Iako model identifikuje ekstremno rizičnu grupu, dodatna podela stabilnih korisnika ne donosi značajnu dodatnu interpretativnu vrednost.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.utils import resample

# =========================================================
# 1. Zadrži samo numeričke kolone (OBAVEZNO)
# =========================================================

df_numeric = df.select_dtypes(include=["int64", "float64"]).copy()

# =========================================================
# 2. Skaliranje podataka
# =========================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_numeric)

# =========================================================
# 3. Testiranje broja klastera (2–8)
# =========================================================

n_range = range(2, 9)
aic_scores = []
bic_scores = []
silhouette_scores = []

# manji uzorak za silhouette (brže računanje)
X_sample = resample(X_scaled, 
                    n_samples=min(5000, len(X_scaled)), 
                    random_state=42)

for n in n_range:
    gmm = GaussianMixture(
        n_components=n,
        random_state=42,
        n_init=5,
        max_iter=200
    )
    
    labels = gmm.fit_predict(X_scaled)

    aic_scores.append(gmm.aic(X_scaled))
    bic_scores.append(gmm.bic(X_scaled))

    # Silhouette na uzorku
    sample_labels = labels[:len(X_sample)]
    silhouette_scores.append(
        silhouette_score(X_sample, sample_labels)
    )

# =========================================================
# 4. Vizualizacija AIC, BIC i Silhouette
# =========================================================

plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.plot(n_range, aic_scores, marker='o')
plt.title("AIC")
plt.xlabel("Broj klastera")

plt.subplot(1,3,2)
plt.plot(n_range, bic_scores, marker='o')
plt.title("BIC")
plt.xlabel("Broj klastera")

plt.subplot(1,3,3)
plt.plot(n_range, silhouette_scores, marker='o')
plt.title("Silhouette")
plt.xlabel("Broj klastera")

plt.tight_layout()
plt.show()

# =========================================================
# 5. Izbor optimalnog broja klastera (po BIC)
# =========================================================

best_n = n_range[np.argmin(bic_scores)]
print("Optimalni broj klastera po BIC:", best_n)

# =========================================================
# 6. Finalni GMM model
# =========================================================

gmm = GaussianMixture(
    n_components=best_n,
    random_state=42,
    n_init=10,
    max_iter=300
)

gmm_labels = gmm.fit_predict(X_scaled)

print("Finalni Silhouette:",
      silhouette_score(X_sample, gmm_labels[:len(X_sample)]))

# =========================================================
# 7. Dodavanje klastera u originalni dataset
# =========================================================

df_with_clusters = df_numeric.copy()
df_with_clusters["Cluster"] = gmm_labels

# =========================================================
# 8. Klaster summary (prosečne vrednosti)
# =========================================================

cluster_summary = (
    df_with_clusters
    .groupby("Cluster")
    .mean()
)

print(cluster_summary)

# =========================================================
# 9. Heatmap klaster profila
# =========================================================

plt.figure(figsize=(12,6))
sns.heatmap(cluster_summary.T,
            cmap="coolwarm",
            annot=True,
            fmt=".2f")

plt.title("Profil klastera (GMM)")
plt.show()
```

## Rezultati GMM klaster analize

Na osnovu BIC kriterijuma, optimalan broj klastera iznosi **8**. Međutim, vrednost Silhouette koeficijenta iznosi **-0.043**, što ukazuje na veoma slabu separaciju klastera i značajno preklapanje između njih. Negativna Silhouette vrednost sugeriše da granice između klastera nisu jasno definisane i da model nema izraženu strukturalnu segmentaciju.

Analiza prosečnih vrednosti po klasterima pokazuje određene razlike u ponašanju korisnika (npr. klaster 6 ima najvišu login frekvenciju, engagement i lifetime value, dok klaster 0 ima nisku aktivnost i 100% churn stopu). Takođe, klasteri 0 i 7 imaju stopu odustajanja od 100%, dok nekoliko klastera (2, 4, 5, 6) ima 0% churn.

Ipak, uprkos tim razlikama, negativan Silhouette rezultat ukazuje da su klasteri formirani više kao statistička posledica modela nego kao jasno razdvojene grupe u prostoru podataka. To znači da GMM, iako identifikuje varijacije u intenzitetu ponašanja korisnika, ne pruža snažnu i stabilnu segmentaciju.

Zaključno, iako BIC sugeriše 8 klastera kao optimalno rešenje sa aspekta verovatnoće modela, kvalitet klastera u smislu separacije nije zadovoljavajući. Ovo ukazuje da GMM u ovom slučaju nema dovoljno izraženu diskriminativnu moć u odnosu na druge metode klasterovanja.
