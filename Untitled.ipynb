## Detekcija i obrada outlier-a

Outlier-i su vrednosti koje značajno odstupaju od ostatka podataka. U kontekstu
klasterizacije, oni mogu imati sledeće negativne efekte:

- distorziraju udaljenost između instanci,
- utiču na centroid-e kod K-Means algoritma,
- mogu dovesti do stvaranja dodatnog klastera koji nije smislen,
- otežavaju interpretaciju rezultata.

Zbog toga je identifikacija i odgovarajuća obrada outlier-a ključna za kvalitet
dobijenih klastera.


Za detekciju outlier-a koristićemo **IQR metodu**, jer je robustna prema
asimetričnim distribucijama, što je slučaj u ovom dataset-u. Z-score metoda
se može koristiti kao dodatna provera.


````python
# Funkcija za detekciju outlier-a
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Provera za najkritičnije numeričke atribute
critical_features = ['Age', 'Average_Order_Value', 'Lifetime_Value', 'Cart_Abandonment_Rate']

for feature in critical_features:
    outliers = detect_outliers_iqr(df_selected, feature)
    print(f"{feature}: {len(outliers)} outlier-a ({len(outliers)/len(df_selected)*100:.2f}%)")
````

IQR analiza pokazuje da su najkritičniji atributi sa najvećim brojem outlier-a:

- *Age* → nekoliko stotina ekstremnih vrednosti (npr. 200 godina)
- *Average_Order_Value* → nekoliko stotina veoma visokih kupovina
- *Lifetime_Value* → ekstremne vrednosti preko 8000
- *Cart_Abandonment_Rate* → vrednosti preko 100%

Ovi outlier-i mogu negativno uticati na klasterizaciju i moraju se obraditi.


Za ovaj projekat odlučeno je da se primeni **capping metoda (winsorizing)**, koja
ograničava ekstremne vrednosti na donju i gornju granicu IQR-a. Na ovaj način
se zadržavaju sve instance, a efekat outlier-a na formiranje klastera se smanjuje.


````python
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# Primena capping-a na kritične feature-e
for feature in critical_features:
    df_selected = cap_outliers_iqr(df_selected, feature)
````

````python
for feature in critical_features:
    outliers = detect_outliers_iqr(df_selected, feature)
    print(f"{feature} after capping: {len(outliers)} outlier-a ({len(outliers)/len(df_selected)*100:.2f}%)")
````

Primena capping metode omogućava očuvanje svih instanci u dataset-u, dok se
istovremeno eliminiše negativni uticaj ekstremnih vrednosti na algoritme
zasnovane na distancama. Ova metoda je posebno pogodna za klasterizaciju i
omogućava dobijanje stabilnijih i interpretabilnih klastera.


## Redukcija dimenzionalnosti

Skup podataka nakon selekcije i kodiranja sadrži više od 20 atributa.
Algoritmi za klasterizaciju, naročito vizuelizacija rezultata, postaju
teško interpretabilni u visoko-dimenzionalnom prostoru.

Redukcija dimenzionalnosti omogućava:

- vizuelizaciju podataka u 2D ili 3D prostoru,
- smanjenje šuma i redundantnih informacija,
- ubrzanje algoritama za klasterizaciju,
- bolju interpretaciju dobijenih klastera.

Najčešće korišćene metode su:
- **PCA (Principal Component Analysis)** – linearna metoda, fokus na varijansu
- **t-SNE (t-Distributed Stochastic Neighbor Embedding)** – nelinearna metoda, fokus na očuvanje lokalnih odnosa


````python
from sklearn.decomposition import PCA

# Kreiramo PCA objekat za 2 komponente
pca = PCA(n_components=2)

# Primena PCA na skalirane podatke
X_pca = pca.fit_transform(X_scaled)

# Pregled oblika podataka
X_pca.shape
````

````python
explained_variance = pca.explained_variance_ratio_
print(f"PC1: {explained_variance[0]*100:.2f}% varijanse")
print(f"PC2: {explained_variance[1]*100:.2f}% varijanse")
````

Prve dve PCA komponente objašnjavaju značajan deo varijanse u podacima.
Ovo nam omogućava da 2D vizualizacijom dobijemo realan pregled strukture
podataka i potencijalnih klastera.


````python
plt.figure(figsize=(10, 7))
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=10, alpha=0.6)
plt.title('2D PCA projection of E-commerce customers')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
````

Vizualizacija PCA pokazuje grupisanje instanci u prostoru prvih dve komponente.
Još ne znamo klastere, ali možemo uočiti potencijalne zone gde se podaci
koncentrišu, što je korisno za inicijalnu procenu broja klastera.


t-SNE je metoda koja bolje očuvava lokalne odnose među podacima, odnosno
slični podaci ostaju bliže, dok se različiti podaci udaljavaju. 
Posebno je korisna za vizualizaciju visokodimenzionalnih podataka.


````python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

tsne = TSNE(n_components=2, perplexity=30, max_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(10, 7))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], s=10, alpha=0.6)
plt.title('2D t-SNE projection of E-commerce customers')
plt.show()
````

t-SNE vizualizacija često otkriva finije strukture u podacima i može pomoći
u izboru broja klastera i identifikaciji potencijalnih izolovanih grupa.


Redukcija dimenzionalnosti omogućava:

- bržu i stabilniju klasterizaciju,
- vizuelnu procenu strukture podataka,
- identifikaciju potencijalnih izolovanih instanci ili “outlier-a”.

PCA je koristan za globalne obrasce, dok t-SNE pomaže u otkrivanju lokalnih
grupa i finijih obrazaca ponašanja korisnika.


## Primena algoritama za klasterizaciju

Korišćenje više algoritama omogućava:

- upoređivanje dobijenih klastera,
- ocenu kvaliteta klastera po različitim kriterijumima,

Najčešće korišćeni algoritmi u ovom projektu:

2. DBSCAN (detekcija gustih regiona i outlier-a)
3. Hijerarhijska klasterizacija (opcioni korak)


````python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []
K_range = range(2, 11)  # testiramo K od 2 do 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Vizualizacija Elbow metode
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, 'bo-')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal K')
plt.show()
````

````python
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    print(f"K={k}, Silhouette Score={score:.4f}")
````

Elbow metoda pokazuje tačku “lakat” gde daljim povećanjem K dobijamo minimalnu
poboljšanje u WCSS. Silhouette score potvrđuje koliko su klasteri dobro odvojeni.
Optimalni K biramo kombinacijom ova dva kriterijuma.


````python
optimal_k = 4  # primer, prema Elbow + Silhouette analizi
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
````

````python
plt.figure(figsize=(10, 7))
for i in range(optimal_k):
    plt.scatter(
        X_pca[kmeans_labels == i, 0],
        X_pca[kmeans_labels == i, 1],
        label=f'Cluster {i}',
        s=20,
        alpha=0.6
    )

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('K-Means Clusters visualized with PCA')
plt.legend()
plt.show()
````

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) grupiše
podatke na osnovu gustine. Glavne prednosti:

- Ne zahteva unapred definisan broj klastera
- Detektuje outlier-e automatski
- Dobro radi kod nelinearnih i nepravilnih oblika klastera


````python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=2, min_samples=10)  # eps = udaljenost, min_samples = gustina
dbscan_labels = dbscan.fit_predict(X_scaled)

# Broj klastera (osim -1 koji su outlier-i)
n_clusters_dbscan = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
print(f"DBSCAN detected {n_clusters_dbscan} clusters")
````

````python
plt.figure(figsize=(10, 7))
unique_labels = set(dbscan_labels)
colors = sns.color_palette(n_colors=len(unique_labels))

for label, color in zip(unique_labels, colors):
    plt.scatter(
        X_pca[dbscan_labels == label, 0],
        X_pca[dbscan_labels == label, 1],
        label=f'Cluster {label}' if label != -1 else 'Outlier',
        c=[color],
        s=20,
        alpha=0.6
    )

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('DBSCAN Clusters visualized with PCA')
plt.legend()
plt.show()
````

DBSCAN je automatski detektovao gustinske klastere i označio outlier-e sa
label-om -1. Ova metoda omogućava uočavanje izolovanih grupa i ekstremnih
instanci koje ne pripadaju ni jednom klasteru.


````python
print(f"K-Means Silhouette Score: {silhouette_score(X_scaled, kmeans_labels):.4f}")

# DBSCAN: samo ako imamo više od 1 klastera
if n_clusters_dbscan > 1:
    print(f"DBSCAN Silhouette Score: {silhouette_score(X_scaled, dbscan_labels):.4f}")
else:
    print("DBSCAN Silhouette Score cannot be computed (only 1 cluster detected)")
````

Silhouette score daje kvantitativnu meru kvaliteta klastera:

- vrednosti bliže 1 → dobri, odvojeni klasteri
- vrednosti oko 0 → preklapanje klastera
- negativne vrednosti → pogrešna klasifikacija

Upoređivanjem K-Means i DBSCAN rezultata možemo odlučiti koji algoritam
najbolje opisuje strukturu podataka.


Primena više algoritama za klasterizaciju omogućava:

- vizualnu i kvantitativnu procenu klastera,
- identifikaciju outlier-a,
- izbor optimalnog algoritma za dalju analizu i interpretaciju.

Kombinacija K-Means i DBSCAN daje potpunu sliku globalnih i lokalnih grupa
u dataset-u korisnika e-commerce platforme.
