## Detekcija i obrada outlier-a

Outlier-i su vrednosti koje značajno odstupaju od ostatka podataka. U kontekstu
klasterizacije, oni mogu imati sledeće negativne efekte:

- distorziraju udaljenost između instanci,
- utiču na centroid-e kod K-Means algoritma,
- mogu dovesti do stvaranja dodatnog klastera koji nije smislen,
- otežavaju interpretaciju rezultata.

Zbog toga je identifikacija i odgovarajuća obrada outlier-a ključna za kvalitet
dobijenih klastera.


Za detekciju outlier-a koristićemo **IQR metodu**, jer je robustna prema
asimetričnim distribucijama, što je slučaj u ovom dataset-u. Z-score metoda
se može koristiti kao dodatna provera.


```python
# Funkcija za detekciju outlier-a
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Provera za najkritičnije numeričke atribute
critical_features = ['Age', 'Average_Order_Value', 'Cart_Abandonment_Rate']

for feature in critical_features:
    outliers = detect_outliers_iqr(df_selected, feature)
    print(f"{feature}: {len(outliers)} outlier-a ({len(outliers)/len(df_selected)*100:.2f}%)")
```

IQR analiza pokazuje da su najkritičniji atributi sa najvećim brojem outlier-a:

- *Age* → nekoliko stotina ekstremnih vrednosti (npr. 200 godina)
- *Average_Order_Value* → nekoliko stotina veoma visokih kupovina
- *Cart_Abandonment_Rate* → vrednosti preko 100%

Ovi outlier-i mogu negativno uticati na klasterizaciju i moraju se obraditi.


Za ovaj projekat odlučeno je da se primeni **capping metoda (winsorizing)**, koja
ograničava ekstremne vrednosti na donju i gornju granicu IQR-a. Na ovaj način
se zadržavaju sve instance, a efekat outlier-a na formiranje klastera se smanjuje.


```python
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# Primena capping-a na kritične feature-e
for feature in critical_features:
    df_selected = cap_outliers_iqr(df_selected, feature)
```

## Kodiranje i skaliranje podataka

Algoritmi za klasterizaciju, kao što su K-Means i DBSCAN, zasnivaju se na
izračunavanju rastojanja između instanci u prostoru atributa. Zbog toga je
neophodno da svi atributi budu numerički i uporedivi po skali.

Kategorijalni atributi moraju biti transformisani u numerički oblik, dok je
numeričke atribute potrebno skalirati kako bi se sprečilo da atributi sa
većim opsegom vrednosti imaju dominantan uticaj na formiranje klastera.


```python
categorical_cols = ['Gender', 'Country', 'Signup_Quarter']
numerical_cols = [col for col in df_selected.columns if col not in categorical_cols]
```

U ovom koraku identifikovani su kategorijalni atributi koji zahtevaju
kodiranje, dok su preostali atributi numeričkog tipa i biće predmet skaliranja.


```python
# ================================
# KOPIJA DATAFRAME-A
# ================================
df_model = df_selected.copy()


# ================================
# DEFINISANJE KOLONA
# ================================
categorical_cols = ['Gender', 'Country', 'Signup_Quarter']
numerical_cols = [col for col in df_model.columns if col not in categorical_cols]


# ================================
# 1️⃣ LABEL ENCODING (Gender)
# ================================
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_model['Gender'] = le.fit_transform(df_model['Gender'])


# ================================
# 2️⃣ TARGET ENCODING (Country)
#    preko broja korisnika iz te zemlje
# ================================
country_freq = df_model['Country'].value_counts(normalize=True)
df_model['Country'] = df_model['Country'].map(country_freq)


# ================================
# 3️⃣ FREQUENCY ENCODING (Signup_Quarter)
# ================================
quarter_freq = df_model['Signup_Quarter'].value_counts(normalize=True)
df_model['Signup_Quarter'] = df_model['Signup_Quarter'].map(quarter_freq)


# ================================
# 4️⃣ STANDARD SCALING
# ================================
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_model)


# ================================
# PROVERA
# ================================
print("Shape nakon encodinga:", df_model.shape)
print("Shape nakon scalinga:", X_scaled.shape)
X_scaled[:5]
```

Smanjenje dimenzionalnosti nakon encodinga

na

23 kolone (kompresovani encoding)

✔ Stabilniji distance space za KMeans

✔ Bolja interpretabilnost klastera

Smanjenje sa 31 → 23 feature-a čini model efikasnijim i pogodnijim za PCA i klaster analizu.

Nakon skaliranja, svi atributi se nalaze na uporedivoj skali, čime je
obezbeđeno da nijedan atribut ne dominira u procesu formiranja klastera.


Kombinacijom odgovarajućeg kodiranja kategorijalnih atributa i skaliranja
numeričkih podataka, formiran je konačni skup podataka spreman za primenu
različitih algoritama za klasterizaciju.

Na osnovu kumulativne varijanse može se videti da je za očuvanje 80% varijanse potrebno 15+ komponenti, dok su dve komponente korišćene isključivo za vizualizaciju.

## Primena algoritama za klasterizaciju

Korišćenje više algoritama omogućava:

- upoređivanje dobijenih klastera,
- ocenu kvaliteta klastera po različitim kriterijumima,
- identifikaciju optimalnog broja klastera,
- razumevanje kako različite metode obrađuju gustinu i distribuciju podataka.

Najčešće korišćeni algoritmi u ovom projektu:

1. K-Means (linearna klasterizacija zasnovana na centroidima)
2. DBSCAN (detekcija gustih regiona i outlier-a)
3. Hijerarhijska klasterizacija (opcioni korak)


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []
K_range = range(2, 11)  # testiramo K od 2 do 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)

# Vizualizacija Elbow metode
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, 'bo-')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import calinski_harabasz_score

K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_pca)
    
    silhouette = silhouette_score(X_pca, labels)
    davies_bouldin = davies_bouldin_score(X_pca, labels)
    calinski = calinski_harabasz_score(X_pca, labels)
    
    print(f"K={k}")
    print(f"  Silhouette Score: {silhouette:.4f}")
    print(f"  Davies-Bouldin Index: {davies_bouldin:.4f}")
    print(f"  Calinski-Harabasz Score: {calinski:.2f}")
    print("-" * 40)
```

Elbow metoda pokazuje tačku “lakat” gde daljim povećanjem K dobijamo minimalnu
poboljšanje u WCSS. Silhouette score potvrđuje koliko su klasteri dobro odvojeni.
Optimalni K biramo kombinacijom ova dva kriterijuma.

Relativno niske vrednosti Silhouette koeficijenta ukazuju na kompleksnu i preklapajuću strukturu podataka, što je čest slučaj kod realnih e-commerce podataka sa velikim brojem atributa.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.utils import resample

# =========================================================
# 1. Zadrži samo numeričke kolone (OBAVEZNO)
# =========================================================

df_numeric = df.select_dtypes(include=["int64", "float64"]).copy()

# =========================================================
# 2. Skaliranje podataka
# =========================================================

scaler = StandardScaler()
X_pca = scaler.fit_transform(df_numeric)

# =========================================================
# 3. Testiranje broja klastera (2–8)
# =========================================================

n_range = range(2, 9)
aic_scores = []
bic_scores = []
silhouette_scores = []

# manji uzorak za silhouette (brže računanje)
X_sample = resample(X_pca, 
                    n_samples=min(5000, len(X_pca)), 
                    random_state=42)

for n in n_range:
    gmm = GaussianMixture(
        n_components=n,
        random_state=42,
        n_init=5,
        max_iter=200
    )
    
    labels = gmm.fit_predict(X_pca)

    aic_scores.append(gmm.aic(X_pca))
    bic_scores.append(gmm.bic(X_pca))

    # Silhouette na uzorku
    sample_labels = labels[:len(X_sample)]
    silhouette_scores.append(
        silhouette_score(X_sample, sample_labels)
    )

# =========================================================
# 4. Vizualizacija AIC, BIC i Silhouette
# =========================================================

plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.plot(n_range, aic_scores, marker='o')
plt.title("AIC")
plt.xlabel("Broj klastera")

plt.subplot(1,3,2)
plt.plot(n_range, bic_scores, marker='o')
plt.title("BIC")
plt.xlabel("Broj klastera")

plt.subplot(1,3,3)
plt.plot(n_range, silhouette_scores, marker='o')
plt.title("Silhouette")
plt.xlabel("Broj klastera")

plt.tight_layout()
plt.show()

# =========================================================
# 5. Izbor optimalnog broja klastera (po BIC)
# =========================================================

best_n = n_range[np.argmin(bic_scores)]
print("Optimalni broj klastera po BIC:", best_n)

# =========================================================
# 6. Finalni GMM model
# =========================================================

gmm = GaussianMixture(
    n_components=best_n,
    random_state=42,
    n_init=10,
    max_iter=300
)

gmm_labels = gmm.fit_predict(X_pca)

print("Finalni Silhouette:",
      silhouette_score(X_sample, gmm_labels[:len(X_sample)]))

# =========================================================
# 7. Dodavanje klastera u originalni dataset
# =========================================================

df_with_clusters = df_numeric.copy()
df_with_clusters["Cluster"] = gmm_labels

# =========================================================
# 8. Klaster summary (prosečne vrednosti)
# =========================================================

cluster_summary = (
    df_with_clusters
    .groupby("Cluster")
    .mean()
)

print(cluster_summary)

# =========================================================
# 9. Heatmap klaster profila
# =========================================================

plt.figure(figsize=(12,6))
sns.heatmap(cluster_summary.T,
            cmap="coolwarm",
            annot=True,
            fmt=".2f")

plt.title("Profil klastera (GMM)")
plt.show()
```
